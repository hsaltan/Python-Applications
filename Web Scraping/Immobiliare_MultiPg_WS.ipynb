```python
import requests, re
from bs4 import BeautifulSoup
import pandas as pd
import numpy
```

```python
pd.set_option('display.max_colwidth', 50)
```

```python
webaddress="http://www.immobiliare.it/Roma/agenzie_immobiliari_provincia-Roma.html"

r = requests.get(webaddress)

c = r.content

soup = BeautifulSoup(c, "html.parser")

page_nr = soup.find("ul",{"class":"pagination pagination__number"})

page_nr = page_nr.find_all("li",{"class":"disabled"})[-1].text

page_nr = int(page_nr)

print(page_nr,"pages were found")
```

```python
company_list = []

base_url="https://www.immobiliare.it/agenzie-immobiliari/roma-provincia/?pag="

for page in range(1, page_nr + 1):    

    r = requests.get(base_url+str(page))
    
    c = r.content
    
    soup = BeautifulSoup(c,"html.parser")
    
    all_content = soup.find_all('ul', class_='agenzie-list')
    
    items = all_content[0].find_all("div", {"class": "block__data"})
    
    for item in items:
        
        try:
        
            companyName = item.find("a").get('title')

            companyAddress = item.find('p').text

            companyAddress = re.sub("\s+", " ", companyAddress)

            companyAddress = re.sub("\-", "", companyAddress)

            company_dictionary = {}

            company_dictionary["Company Name"]= companyName

            company_dictionary["Company Address"]= companyAddress

            company_list.append(company_dictionary)
            
        except Exception:
            
            pass
```

```python
df = pd.DataFrame(company_list)
df = df.head(10)
df.style  
df.style.set_properties(**{'background-color': 'black',
                           'color': 'lawngreen',
                           'border-color': 'white', 'text-align': 'left'})
```
