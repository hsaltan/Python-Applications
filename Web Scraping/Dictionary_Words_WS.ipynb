```python
import requests, re
from bs4 import BeautifulSoup
import pandas as pd
```

```python
word_list = []

base_url="http://www.dictionary.com/browse/"

words = ["soft", "handy", "lovely", "scrape"]

for word in words:

    word_dictionary = {} 

    r=requests.get(base_url + word)
    
    c=r.content
    
    soup=BeautifulSoup(c,"html.parser")
    
    all_content=soup.find("div", {"class": "default-content"})
    
    all_content=soup.find_all("span", {"class": "one-click-content css-nnyc96 e1q3nk1v1"})
     
    lines = [line.text for line in all_content if not line('span')]
    
    for ind, line in enumerate(lines):
            
        word_dictionary[word.upper() + " def. " + str(ind+1)] = line

    word_list.append(word_dictionary)
```

```python
list_dataframe = []

for data in word_list:

    list_dataframe.append(pd.DataFrame(data.items(), columns = ['Word','Definition']))
    
result_dataframe = pd.concat(list_dataframe, ignore_index=True)

result_dataframe

df = result_dataframe.style.set_properties(**{'text-align': 'left'})

df
```
